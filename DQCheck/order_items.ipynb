```python
import pandas as pd
order_items_df = pd.read_csv('../Landing/order_items/order_items.csv')
order_items_df
```

```python
#check nulls
order_items_df.isnull().sum()
```

```python
order_items_df = order_items_df[(order_items_df['list_price'] >= 200) & (order_items_df['list_price'] <= 10000)]
```

```python
max_price=order_items_df['list_price'].max()
max_price
```

```python
min_price=order_items_df['list_price'].min()
min_price
```

```python
#check dtype of item_id
order_items_df.info()
```

```python
# Convert 'item_id' column to numeric, coerce non-numeric values to NaN
order_items_df['item_id'] = pd.to_numeric(order_items_df['item_id'], errors='coerce')
# Drop rows with NaN values in 'item_id' column
order_items_df.dropna(subset=['item_id'], inplace=True)
```

```python
order_items_df.isnull().sum()
```

```python
order_items_df['item_id'] = order_items_df['item_id'].astype(int)
```

```python
order_items_df.info()
```

```python
products= pd.read_csv('../Landing/products/products.csv')
products
```

```python
order_items_in_products = order_items_df['item_id'].isin(products['product_id'])
order_items_in_products
```

```python
order_items_df = order_items_df[order_items_in_products]
```

```python
order_items_df
```
