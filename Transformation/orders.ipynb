```python
import pandas as pd 
orders=pd.read_csv('../Staging_1/orders/orders.csv')
orders.info()
```

```python
orders['required_date'] = pd.to_datetime(orders['required_date'], errors='coerce')
orders['shipped_date'] = pd.to_datetime(orders['shipped_date'], errors='coerce')
orders['late_delivery'] = orders.apply(lambda row: (row['shipped_date'] - row['required_date']).days > 0, axis=1)
```

```python
orders
```

```python
orders['latency_days'] = orders.apply(lambda row: (row['shipped_date'] - row['required_date']).days if row['late_delivery'] else None, axis=1)
orders
```

```python
orders['weekend'] = orders['shipped_date'].apply(lambda x: x.weekday() in [6, 0])
orders
```

```python
order_status_dict= {
    'status_code': [1, 2, 3, 4],
    'status_description': ['Pending', 'Processing', 'Rejected', 'Completed']
}
order_status_lookup = pd.DataFrame(order_status_dict)
order_status_lookup.to_csv('../Staging_2/order_status_lookup/order_status_lookup.csv', index=False)
```
